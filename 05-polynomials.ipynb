### 1 (computer problems)

Write a program that takes as input the coefficients of a polynomial $p$ and a specific point $z_0$ and produces as output the values $p(z_0)$, $p'(z_0)$, and $p''(z_0)$. Write the pseudocode with only one loop. Test on the polynomial of the first no computer problem, taking $z_0 = 4$.

#### Not using a single loop

````python
def derivate(coefficients: list[int], derivate_number: int) -> list[int]:
	for _ in range(derivate_number):
		n = len(coefficients)
		output = []

		for i in range(1, n + 1):
			exponent = n - i
			coefficient = coefficients[i - 1]
			output.append(coefficient * exponent)

		output.pop()

		coefficients = output

	return output
````

````python
p = [3, -7, -5, 1, -8, 2]
p_1 = derivate(p, 1)
p_2 = derivate(p, 2)

print(p_1, p_2)
````

````python
def Horner_algorithm(p_coefficients: list[int], z_0: int) -> int:
	n = len(p_coefficients)
	b = [0] * n

	b[0] = p_coefficients[0]

	for i in range(1, n):
		b[i] = p_coefficients[i] + z_0 * b[i - 1]

	# n = len(p_coefficients) - 1
	# b = [0] * (n + 1)

	# b[n] = p_coefficients[n]

	# for i in range(n - 1, -1, -1):
	# 	b[i] = p_coefficients[i] + z_0 * b[i + 1]

	return b[n - 1]
````

````python
p_result = Horner_algorithm(p, 4)
p_1_result = Horner_algorithm(p_1, 4)
p_2_result = Horner_algorithm(p_2, 4)

print(p_result, p_1_result, p_2_result)
````

#### Using a single loop

Let's consider a general polynomial of degree $n$:

$$
p(x) = a_n x^n + a_{ n - 1 } x^{ n - 1 } + \dots + a_1 x + a_0
$$

Then the first derivative corresponds to:

$$
p'(x) = n a_n x^{ n - 1 } + (n - 1) a_{ n - 1 } x^{ n - 2 } + \dots + a_1
$$

And finally the second derivative is:

$$
p''(x) = n (n - 1) a_n x^{ n - 2 } + (n - 1) (n - 2) a_{ n - 1 } x^{ n - 3 } + \dots + a_2
$$

```
input:
    - Coefficients of the polynomial: coefficients[0...n] (where coefficients[i] is the coefficient of x^i)
    - A specific point: z_0

output:
    - p(z_0), p'(z_0), p''(z_0)

p_val -> 0  # p(z_0)
p_1_val -> 0  # p'(z_0)
p_2_val -> 0  # p''(z_0)
n -> length of coefficients - 1  # degree of the polynomial

for i from n to 0 do
    p_val = p_val * z_0 + coefficients[i]

    if i > 0 then
        p_1_val = p_1_val * z_0 + coefficients[i] * i

    if i > 1 then
        p_2_val = p_2_val * z_0 + coefficients[i] * i * (i - 1)
end do

output p_val, p_1_val, p_2_val
```

### 4  (computer problems)

Using Newton's method and the polynomial $p(z) = z^3 - 1$, find three nearby starting points (within $0.01$ of each Other) so that the resulting sequences converge to different roots. Using a plotter, show the paths of these sequences of points within a square containing the roots by connecting successive points with line segments.

> The paths refer to the points through which Newton's method passes.


````python
colors_dictionary = {
	"black": "#1C1C1C",
	"red": "#AF5F5F",
	"green": "#5F875F",
	"yellow": "#87875F",
	"blue": "#5F87AF",
	"magenta": "#5F5F87",
	"cyan": "#5F8787",
	"white": "#6C6C6C",
	"bright_black": "#444444",
	"bright_red": "#D78787",
	"bright_green": "#87AF87",
	"bright_yellow": "#FFFFAF",
	"bright_blue": "#87AFD7",
	"bright_magenta": "#8787AF",
	"bright_cyan": "#5FAFAF",
	"bright_white": "#FFFFFF"
}
````

````python
def newton_method(f, f_prime, x0, max_iteration=300, tolerance=1e-6):
	x_old = x0
	points = [x_old]
	for _ in range(max_iteration):
		f_value = f(x_old)
		f_prime_value = f_prime(x_old)

		if f_prime_value == 0:
			raise ValueError("Error: derivative is zero")

		x_new = x_old - f_value / f_prime_value
		points.append(x_new)

		if abs(x_new - x_old) < tolerance:
			break

		x_old = x_new

	return points
````

````python
def p(z):
	return z ** 3 - 1

def p_prime(z):
	return 3 * z ** 2
````

````python
import numpy as np
import matplotlib.pyplot as plt


z0 = 0.5 + 0.5j
delta = 0.005
initial_points = [z0, z0 + delta, z0 + delta*1j]

colors = [colors_dictionary["red"], colors_dictionary["green"], colors_dictionary["blue"]]

roots = [1, -1/2 + np.cbrt(3)/ 2j, -1/2 - np.cbrt(3)/ 2j]
plt.plot([root.real for root in roots], [root.imag for root in roots], "ko", label="roots")

for i, z_initial in enumerate(initial_points):
	points = newton_method(p, p_prime, z_initial)
	x_vals = [z.real for z in points]
	y_vals = [z.imag for z in points]
	plt.plot(x_vals, y_vals, marker="o", color=colors[i], label=f"path { i + 1 }")

plt.legend()
plt.xlabel("Re(z)")
plt.ylabel("Im(z)")
plt.title("path of Newton's method ($z^3 - 1$)")
plt.grid(True)
plt.show()
````
