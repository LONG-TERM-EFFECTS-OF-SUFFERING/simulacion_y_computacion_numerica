## Usage example

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
````

### Drawing a sine graph

````python
plt.xlabel("x")
plt.ylabel("y")
plt.title("sin")
plt.plot(x, y_sin)
````

### Drawing a sin and cos graph

````python
plt.xlabel("x")
plt.ylabel("y")
plt.title("sin")
plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.legend(["sin", "cos"])
````

### Drawing a sin and cos graph individually

````python
plt.subplot(2, 1, 1)
plt.plot(x, y_sin)
plt.title("sin")

plt.subplot(2, 1, 2)
plt.plot(x, y_cos)
plt.title("cos")

plt.show()
````
